```python
import serial
import time

# === Configuration ===
SERIAL_PORT = '/dev/tty.usbserial-10'  # Replace with your serial port (e.g., 'COM3' for Windows, '/dev/ttyUSB0' for Linux)
BAUD_RATE = 9600

# === Initialize Serial Connection ===
try:
    ser = serial.Serial(SERIAL_PORT, BAUD_RATE, timeout=1)
    print(f"Connected to {SERIAL_PORT} at {BAUD_RATE} baud.")
except serial.SerialException as e:
    print(f"Error: {e}")
    exit()

# === Helper Functions ===
def send_command(command):
    """
    Sends a command to the robot over serial and waits for a response.
    """
    if ser.is_open:
        ser.write((command + '\n').encode())  # Send command as bytes
        time.sleep(0.1)  # Wait for a short delay
        while ser.in_waiting > 0:  # Check for responses
            response = ser.readline().decode().strip()
            print(f"Robot Response: {response}")
    else:
        print("Serial connection is not open.")

# === Commands ===
def rotate(angle):
    """
    Rotates the robot by a specific angle (positive = right, negative = left).
    """
    send_command(f"rotate,{angle}")

def move_forward(distance):
    """
    Moves the robot forward by a specified distance in mm.
    """
    send_command(f"forward,{distance}")

def grid_move(x, y):
    """
    Moves the robot to a specified grid coordinate (x, y).
    """
    send_command(f"grid,{x},{y}")


```
