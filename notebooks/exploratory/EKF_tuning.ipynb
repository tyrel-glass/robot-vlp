```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import robot_vlp.data_collection.experment_processing as ep
import robot_vlp.modeling.gen_cnc_vlp_model as vlp
import robot_vlp.data_collection.communication as c
import pickle

from robot_vlp.config import EXPERIMENT_DATA_DIR, VLP_MODELS_DIR


%load_ext autoreload
%autoreload 2
```

```python
vlp_models = vlp.load_vlp_models()
vlp_model = vlp_models['high_acc']
df_lst = []
for i in range(3,13):
    path = EXPERIMENT_DATA_DIR/f'Robot/exp1_{i}.csv'
    df_lst.append(ep.process_robot_exp_file(path, vlp_model))
```

```python
import robot_vlp.modeling.EKF as ekf
```

```python
optimized_params, result, optimization_progress = ekf.tune_ekf(df_lst[3:-1], initial_guess = [1.0, 1.0, 1.0])
```

```python
# Extract iterations and function values
iterations = [entry[0] for entry in optimization_progress]
errors = [entry[1] for entry in optimization_progress]

# Plot progress
plt.figure(figsize=(8, 5))
plt.plot(iterations, errors, marker="o", linestyle="-", color="b")
plt.xlabel("Iteration")
plt.ylabel("Best Error (f(x))")
plt.title("Convergence of Differential Evolution")
plt.grid(True)
plt.show()
```
